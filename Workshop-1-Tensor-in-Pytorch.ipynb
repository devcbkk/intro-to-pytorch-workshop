## NumPy

```
import numpy as np
```

```
a = np.array([[1, 2, 3],
              [4, 5, 6]], dtype=np.float)

print(a)
```

```
b = np.array(2, dtype=np.float)
print(b)
```

```
c = a + b
print(c)
```

```
d = c[:, [0, 2]]
print(d)
```

```
assert d.shape == (2, 2)
```

```
np.random.seed(42)

r = np.random.normal(size=(2, 2))
print(r)
```

```
e = d @ r
print(e)
```

## PyTorch as NumPy

```
import torch
```

```
a = torch.tensor([[1, 2, 3],
                  [4, 5, 6]], dtype=torch.float)

print(a)
```

```
b = torch.tensor(2, dtype=torch.float)

print(b)
```

```
c = a + 2
print(c)
```

```
d = c[:, [0, 2]]
print(d)
```

```
assert d.shape == (2, 2)
```

```
torch.manual_seed(42)

r = torch.rand(2, 2, dtype=torch.float)
print(r)
```

```
e = d @ r
print(e)
```

## Tensor on GPU

```
dev = 'cuda:0' if torch.cuda.is_available() else 'cpu'
print(dev)
```

```
x = torch.randn((2, 2), device=dev)
print(x)
```

```
y = torch.randn((2, 2)).to(x.device)
print(y)
```

```
z = x + y
print(z)
```

## More About Tensor

```
print(z.type())
```

```
z = z.to(torch.float64)
print(z)
```

```
z = z.to(torch.device('cpu'))
print(z.device)
```

```
print(z.numpy())
print(type(z.numpy()))
```

```
print(torch.from_numpy(z.numpy()))
```
